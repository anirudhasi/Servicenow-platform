# Tower & SDM Comparative Analytics

## Purpose
Benchmark organizational performance across towers and service delivery managers.

## Key Metrics by Tower/SDM

### 1. Incident Volume
```
Volume = COUNT(incidents) by tower/sdm
Volume_per_Agent = Volume / Number_of_agents
Trend_YoY = (Current_Year - Previous_Year) / Previous_Year * 100
```

### 2. SLA Compliance %
```
Compliance = (COUNT(made_sla=TRUE) / Total) * 100 by tower/sdm
Target: 90%+
Risk Threshold: <85% triggers intervention
```

### 3. Average MTTR (Hours)
```
MTTR = AVG(mttr_hours) by tower/sdm
Benchmark: Identify leader (lowest MTTR)
Compare to targets: P1=4h, P2=8h, P3=72bh, P4=120bh
```

### 4. Performance Score (Composite)
```
Score = (0.4 * SLA_Compliance) + (0.3 * (100 - MTTR_Variance)) + (0.2 * FAR_metric) + (0.1 * Volume_growth)
Range: 0-100
Green: 80+
Amber: 70-79
Red: <70
```

### 5. First Assignment Rate (FAR)
```
FAR = (COUNT(reassignment_count=0) / Total) * 100
Target: >95% (minimize rework)
Impact: Each reassignment adds 4-12h to MTTR
```

## Comparative Analysis Methods

### A/B Comparison Matrix
- Tower vs SDM: Which tower has best-performing SDMs?
- Service offering: Which group excels at specific categories?
- Priority handling: Who resolves P1/P2 fastest?

### Radar Chart (5-Dimension Comparison)
1. SLA Compliance %
2. Efficiency (Inverse MTTR)
3. Quality (Low reopen rate)
4. Capacity (Revenue per team member)
5. Growth (YoY improvement)

### Pareto Analysis (80/20 Rule)
- Identify top 20% of towers handling 80% of incidents
- Identify top performers for mentoring/best practice sharing


````python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Get filter options to list towers/SDMs
API_BASE = 'http://127.0.0.1:8002/api'

response = requests.get(f'{API_BASE}/monitoring/filters')
filters = response.json()

towers = filters.get('towers', [])
sdms = filters.get('sdms', [])

print(f'Towers: {towers}')
print(f'SDMs: {sdms}')
````
