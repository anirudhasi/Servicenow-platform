# M2 Trend Analysis — Historical Patterns & Forecasting

## Purpose
Time-series analysis of incident trends over 6 months, with forecasting and pattern detection.

## Methodology

### 1. Volume Trend Analysis
**Formula**:
```
Volume(date) = COUNT(incidents created on date)
Trend = ROLLING_AVG(Volume, 7 days)
YoY Growth = (Current Month - Previous Month) / Previous Month * 100
```

**Seasonality Detection**:
- Day-of-week effect (weekends vs weekdays)
- Month-end/year-end peaks (month-close activities)
- Holiday impact analysis

### 2. MTTR Trend (Mean Time To Resolution)
**Formula**:
```
MTTR(date) = AVG(mttr_hours) for incidents resolved on date
Target by Priority: P1=4h, P2=8h, P3=72bh, P4=120bh
Variance = STDDEV(MTTR)
Trend Direction = Linear regression slope
```

**By-Group Comparison**:
- Identify groups improving vs degrading
- Benchmark against organizational average

### 3. SLA Compliance Trend
**Formula**:
```
Compliance%(date) = (Compliant / Total) * 100 for incidents closed on date
Target: 90% minimum
Breach Risk Window: Track approaching 75%, 50% compliance
```

### 4. Category Distribution Over Time
**Chart Type**: Stacked area chart
**Methodology**:
```
Category_Volume(date) = COUNT(incidents with category created on date)
Stacked % = Category_Volume / Total_Volume * 100
```

### 5. Priority Trend
**Chart Type**: Line chart, one per priority
**Methodology**: COUNT by priority per day, smoothed with 7-day MA

### 6. Reassignment Analysis
**Formula**:
```
Reassignment_Rate = COUNT(reassignment_count > 0) / Total
Avg_Reassignments_per_Incident = AVG(reassignment_count)
Impact_on_MTTR = MTTR(reassignment_count=0) vs MTTR(reassignment_count>0)
```

### 7. Forecast Model (6-period ahead)
**Approach**: ARIMA(1,1,1) for volume forecasting
```
Model: Auto-ARIMA on 6-month volume history
Confidence Intervals: 95% bounds
Seasonality: Weekly pattern captured
```

## Filters
- **Granularity**: day, week, month
- **Tower/SDM/Group**: Multi-select
- **Priority**: Multi-select
- **Date Range**: Last 6 months default

## Visualization Best Practices
1. **Volume**: Stacked bar + line trend overlay
2. **MTTR**: Line chart with target reference line
3. **SLA Compliance**: Area fill (above 90% = green)
4. **Forecast**: Solid line (historical) + dashed (forecast) with confidence band
